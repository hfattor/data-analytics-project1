# Statistical Testing: Correlation
### How are the two variables of sightings per year and population density correlated for low and high density states?

```python
#Using ufohigh_df for high density states:
#Series of # sightings per year
high_counts = ufohigh_df["year"].value_counts()
high_counts = high_counts.to_frame().reset_index()
high_counts.rename(columns={"year":"counts","index":"year"}, inplace=True)
density_counts_df = pd.merge(high_counts, ufohigh_df[['year','decade','Resident Population Density']],
                             on="year",how="left")
density_counts_df.head()
```

```python
density_counts_df.corr(method="pearson")
```

```python
#Here we want to look for the coefficient between counts and Resident Population density: 0.117773.
#In other words, there is very weak positive correlation between the counts and Resident Population Density for high density states.
#Now for low density states:
low_counts = ufolow_df["year"].value_counts()
low_counts = low_counts.to_frame().reset_index()
low_counts.rename(columns={"year":"counts","index":"year"}, inplace=True)
lowden_counts_df = pd.merge(low_counts, ufolow_df[['year','decade','Resident Population Density']],
                             on="year",how="left")
lowden_counts_df.head()
```

```python
lowden_counts_df.corr(method="pearson")
```

```python
#Similar to our first findings, though slightly stronger, our correlation coefficient is 0.263391, still little but positive.
#Low density states have slightly higher correlation in ufo sightings by decade than high density states.
#But with such a low r-value, we can't say much more!
#Graphing the regression line for each density:
stats.pearsonr(density_counts_df["counts"],density_counts_df["Resident Population Density"])
(h_slope,h_intercept,h_rvalue,h_pvalue,h_stderr)=stats.linregress(density_counts_df["counts"],
                                                  density_counts_df["Resident Population Density"])
plt.figure(figsize=(14,10))
for state in ufohigh_df["state"].unique():
    state_df = ufohigh_df.loc[ufohigh_df["state"]==state]
    plt.scatter(state_df["year"],state_df["Resident Population Density"], 
                color = colors[np.where(ufohigh_df["state"].unique() == state)[0][0]],
                marker = 'o', alpha=0.05)
    plt.legend(ufohigh_df["state"].unique(), loc="upper left")
x = range(1950,2014)
y = x*h_slope + h_intercept
plt.text(1990,100, "y = " + str(round(h_slope,4)) +"x + " + str(round(h_intercept, 1)), fontsize = "x-large", color = "b")
plt.plot(x,y)
plt.title("Regression Equation for UFO Sightings per Year vs. Population Density for High Density States")
plt.xlabel("Year and Decade")
plt.ylabel("Population Density")
print("The r-value is " + str(h_rvalue).)
plt.savefig("Output/highdensity_regression.png")
```

```python
stats.pearsonr(lowden_counts_df["counts"],lowden_counts_df["Resident Population Density"])
(l_slope,l_intercept,l_rvalue,l_pvalue,l_stderr)=stats.linregress(lowden_counts_df["counts"],
                                                  lowden_counts_df["Resident Population Density"])
plt.figure(figsize=(14,10))
for state in ufolow_df["state"].unique():
    state_df = ufolow_df.loc[ufolow_df["state"]==state]
    plt.scatter(state_df["year"],state_df["Resident Population Density"], 
                color = colors[np.where(ufolow_df["state"].unique() == state)[0][0]],
                marker = 'o', alpha=0.05)
    plt.legend(ufolow_df["state"].unique(), loc="upper left")
x_l = range(1950,2014)
y_l = x_l*l_slope + l_intercept
plt.text(1990,20, "y = " + str(round(l_slope,4)) +"x + " + str(round(l_intercept, 1)), fontsize = "x-large", color = "b")
plt.plot(x_l,y_l)
plt.title("Regression Equation for UFO Sightings per Year vs. Population Density for Low Density States")
plt.xlabel("Year and Decade")
plt.ylabel("Population Density")
print(x_l[0],l_slope,l_intercept)
print("The r-value is " + str(l_rvalue)+".")
plt.savefig("Output/lowdensity_regression.png")
```
