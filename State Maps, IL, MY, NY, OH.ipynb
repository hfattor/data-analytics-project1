```python
 # Import the required libraries
import hvplot.pandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Turn off warning messages
import warnings
warnings.filterwarnings("ignore")
```

```python
file_one = "output/ufo_statepop_df.csv"
```

```python
ufo_statepop_df = pd.read_csv(file_one)
ufo_statepop_df
```

```python
ufo_ill = ufo_statepop_df.loc[ufo_statepop_df["state"]=="Illinois"]
ufo_ill
```

```python
map_plot_1 =ufo_ill.hvplot.points(
    "longitude ",
    "latitude",
    geo = True,
    tiles = "OSM",
    frame_width=1000,
    frame_height=714,
    size = "Resident Population Density",
    color = "decade",
    alpha = 0.10
    )

# Display the map
map_plot_1
```

```python
ufo_ill.plot(kind="scatter", x="latitude", y="Resident Population Density", grid=True, figsize=(8,8),
              title="latitude vs Resident Population Density")
plt.show()
```
