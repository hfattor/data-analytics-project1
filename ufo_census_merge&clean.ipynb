# UFO Data
Binning UFO data into decades, pulling data for 11 focus states.

```python
ufo_dropped = ufo_raw.dropna()

#datetime
ufo_us = ufo_dropped[ufo_dropped["country"] == "us"]
ufo_by_year = ufo_us
ufo_by_year["datetime"] = pd.to_datetime(ufo_by_year["datetime"], format="%m/%d/%Y %H:%M", errors='coerce')
ufo_by_year[ufo_by_year["datetime"].isna()].dropna()
ufo_by_year["year"] = pd.DatetimeIndex(ufo_by_year['datetime']).year
ufo_by_year.head()

#grouping by decade
groupnames1 = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
bins = [1950.0, 1959.0, 1969.0, 1979.0,1989.0,1999.0,2009.0,2019,2024.0]
ufo_by_year["decade"] = pd.cut(ufo_by_year["year"],bins,labels=groupnames1,include_lowest=False)
```

```python
ufo_by_year = ufo_by_year.dropna()
ufo_by_year
```

```python
columns = ['datetime', 'city', 'state', 'country', 'shape', 'duration (seconds)',
       'duration (hours/min)', 'latitude', 'longitude ', 'year', 'decade']

spotted_in_us_df = ufo_by_year.loc[(ufo_by_year["state"] == "az") | (ufo_by_year["state"] == "ca")| (ufo_by_year["state"] == "fl") 
                                   | (ufo_by_year["state"] == "il") | (ufo_by_year["state"] == "mi") 
                                   | (ufo_by_year["state"] == "ny") | (ufo_by_year["state"] == "oh") | (ufo_by_year["state"] == "or")
                                   | (ufo_by_year["state"] == "pa") | (ufo_by_year["state"] == "tx")
                                   | (ufo_by_year["state"] == "wa"), columns]
```

```python
states = {'az': 'Arizona', 'ca': 'California', 'fl': "Florida", 'il': "Illinois", 'mi': "Michigan", 'ny': "New York", 
          'oh': "Ohio", 'or': "Oregon", 'pa': "Pennsylvania", 'tx': "Texas", 'wa': "Washington"}
states = {state: abbrev for state, abbrev in states.items()}
spotted_in_us_df['state'] = spotted_in_us_df['state'].map(states)

spotted_in_us_df
```

# Merging UFO and State Census Data

```python
ufo_statepop_df = pd.merge(spotted_in_us_df, state_pop_df, 
                                 how='left', on=['decade', 'state'], sort=False)
ufo_statepop_df
```
